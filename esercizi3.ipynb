## Esercizio 4 - `@staticmethod`

Nella classe `Ruolo`, scrivi il metodo statico `normalizza(testo)`
che restituisce il testo **senza spazi ai lati** e in **minuscolo**
(es. `'  UTENTE '` -> `'utente'`). Non deve usare `self`.

```python
# TODO: completa il metodo statico
class Ruolo:
    @staticmethod
    def normalizza(testo):
        return testo.lower().strip()
```

```python
# Self-test
assert Ruolo.normalizza('  UTENTE ') == 'utente'
assert Ruolo.normalizza('Assistant') == 'assistant'
print('OK')
print(Ruolo.normalizza('  SYSTEM '))
```

## Esercizio 5 - `@classmethod` factory

La classe `Messaggio` ha `__init__(self, ruolo, contenuto)`. Aggiungi
due **classmethod** factory:
- `da_utente(testo)` -> crea un messaggio con ruolo `'user'`,
- `da_sistema(testo)` -> crea un messaggio con ruolo `'system'`.

Usa `cls(...)`, non il nome della classe.

```python
# TODO: completa i due classmethod
class Messaggio:
    def __init__(self, ruolo, contenuto):
        self.ruolo = ruolo
        self.contenuto = contenuto

    @classmethod 
    def da_utente(cls, testo):
        return cls('user', testo)

    @classmethod 
    def da_sistema(cls, testo):
        return cls('system', testo)
```

```python
# Self-test
u = Messaggio.da_utente('Ciao')
s = Messaggio.da_sistema('Sei utile')  
assert u.ruolo == 'user' and u.contenuto == 'Ciao'
assert s.ruolo == 'system' and s.contenuto == 'Sei utile'
print('OK')
print(u.ruolo, '|', s.ruolo)
```
