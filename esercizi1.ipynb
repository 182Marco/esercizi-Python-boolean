## Esercizio 1 — F-strings: riga di report

Costruisci la stringa `riga` che descrive un modello LLM, usando **una sola f-string**.

Dati i valori sottostanti, `riga` deve risultare **esattamente**:

```
claude-opus-4.7 | costo: $15.0000 | token: 1,234,567
```

Suggerimenti:
- il costo va mostrato con **4 decimali** (`:.4f`);
- i token con il **separatore delle migliaia** (`:,`).

````python
modello = "claude-opus-4.7"
costo = 15.0
token = 1234567

# TODO: costruisci 'riga' con una f-string (vedi formato richiesto sopra)
riga = f"{modello} | costo: ${costo:.4f} | token: {token:,}"  # <-- sostituisci

print(riga)
````

````python
# Self-test (eseguilo dopo aver completato il TODO sopra)
try:
    assert riga == "claude-opus-4.7 | costo: $15.0000 | token: 1,234,567", f"ottenuto: {riga!r}"
    print("Esercizio 1: OK ✅")
except Exception as e:
    print("Esercizio 1: ❌", type(e).__name__, "->", e)
````

## Esercizio 2 — List comprehension: filtra e ordina

Data la lista `lunghezze_prompt`, crea la lista `lunghe` contenente **solo** i valori
**maggiori di 100**, in **ordine crescente**.

Usa una **list comprehension** (eventualmente combinata con `sorted(...)`).

Risultato atteso: `[450, 600, 1024]`

````python
lunghezze_prompt = [12, 450, 3, 1024, 88, 2, 600]

# TODO: filtra i valori > 100 con una list comprehension, poi ordina in modo crescente
lunghe = sorted([n for n in lunghezze_prompt  if n > 100])  # <-- sostituisci

print(lunghe)
````

## Esercizio 3 — Dict comprehension

Data la lista `parole`, crea il dizionario `mappa` che associa **ogni parola alla sua lunghezza**,
ma **solo** per le parole con **almeno 3 caratteri** (lunghezza ≥ 3).

Usa una **dict comprehension**.

Risultato atteso: `{'machine': 7, 'learning': 8, 'LLM': 3, 'rag': 3}`

````python
parole = ["AI", "machine", "learning", "LLM", "rag"]

# TODO: dict comprehension parola -> len(parola), solo se len >= 3
mappa = {p: len(p) for p in parole if len(p) >= 3}  # <-- sostituisci

print(mappa)
````

````python
# Self-test (eseguilo dopo aver completato il TODO sopra)
try:
    assert mappa == {"machine": 7, "learning": 8, "LLM": 3, "rag": 3}, f"ottenuto: {mappa!r}"
    print("Esercizio 3: OK ✅")
except Exception as e:
    print("Esercizio 3: ❌", type(e).__name__, "->", e)
````

## Esercizio 5 — Type hints: funzione `media`

Completa la funzione `media` in modo che:

- sia **annotata con i type hints**: prende una `list[float]` e restituisce un `float`;
- restituisca la **media aritmetica** dei valori;
- se la lista è **vuota**, restituisca `0.0` (evita la divisione per zero).

I test verificano sia il **risultato** sia la **presenza delle annotazioni**.

````python
# TODO: aggiungi i type hints alla firma e completa il corpo
def media(valori: list[float]) -> float:
    if not valori:
      return 0.0
    return sum(valori) / len(valori);

# Dopo aver completato la funzione, esegui la cella "Self-test" qui sotto.
````

````python
# Self-test (eseguilo dopo aver completato il TODO sopra)
try:
    assert media([2.0, 4.0, 6.0]) == 4.0, "media errata"
    assert media([]) == 0.0, "il caso lista vuota deve restituire 0.0"
    assert media.__annotations__.get("valori") == list[float], "manca/errato il type hint su valori"
    assert media.__annotations__.get("return") == float, "manca/errato il type hint sul valore di ritorno"
    print("Esercizio 5: OK ✅")
except Exception as e:
    print("Esercizio 5: ❌", type(e).__name__, "->", e)
````

## Esercizio 6 — Mini-progetto (combinato)

Mette insieme **tutto**: comprehension + f-strings + type hints.

Scrivi la funzione **annotata** `report(modelli, token, budget)` che:

1. considera **solo** i modelli con `costo_input_mtok` **≤ `budget`** (filtro);
2. per ognuno calcola il costo: `costo = costo_input_mtok * token / 1_000_000`;
3. **ordina** i modelli per costo **crescente**;
4. restituisce una **stringa multilinea** con una riga per modello, formattata così
   (nome largo 8 a sinistra, costo con 4 decimali):
   `f"{nome:<8}| ${costo:.4f}"`, righe unite da `"\n"`.

Con `token=500_000` e `budget=3.0`, il risultato atteso è:

```
haiku   | $0.5000
sonnet  | $1.5000
```
(il modello `opus` è escluso perché costa più del budget).

````python
from ast import Invert
from typing import TypedDict, List

class Modello(TypedDict):
    nome: str
    costo_input_mtok: float




modelli: list[Modello] = [
    {"nome": "haiku",  "costo_input_mtok": 1.0},
    {"nome": "sonnet", "costo_input_mtok": 3.0},
    {"nome": "opus",   "costo_input_mtok": 15.0},
]

token=500_000
budget=3.0

# TODO: completa la funzione (aggiungi anche i type hints!)
def report(modelli: list[Modello], token: int, budget: float) -> str:
    modelliInBudget = [m for m in modelli if m["costo_input_mtok"] <= budget ]

    for item in modelliInBudget:
        item["costo"] = item["costo_input_mtok"] * token / 1_000_000;

    modelliInBudget.sort(key = lambda m: m["costo"])

    for item in modelliInBudget:
        item["costo"] = item["costo"]

    res = '';
    for item in modelliInBudget:
        res += f"{item['nome']:<8}| ${item['costo']:.4f}\n"

    print(res);

    return res.strip()


report(modelli, token, budget)



# Dopo aver completato la funzione, esegui la cella "Self-test" qui sotto.
````

````python
# Self-test (eseguilo dopo aver completato il TODO sopra)
try:
    atteso = "haiku   | $0.5000\nsonnet  | $1.5000"
    ottenuto = report(modelli, 500_000, 3.0)
    assert ottenuto == atteso, f"ottenuto:\n{ottenuto!r}\natteso:\n{atteso!r}"
    assert report.__annotations__.get("return") == str, "manca il type hint sul ritorno (str)"
    print("Esercizio 6: OK ✅")
except Exception as e:
    print("Esercizio 6: ❌", type(e).__name__, "->", e)
````
