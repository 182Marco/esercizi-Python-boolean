## Esercizio 1 - Ereditarieta' base

C'e' una classe `Animale` con `__init__(self, nome)`. Scrivi la
sottoclasse `Gatto` che eredita da `Animale` e aggiunge il metodo
`verso()` che restituisce la stringa `'Miao'`.

```python
class Animale:
    def __init__(self, nome):
        self.nome = nome

# TODO: completa la sottoclasse Gatto
class Gatto(Animale):
    def __init__(self,nome):
        super().__init__(nome)
        

    def verso(self):
        return 'Miao'
```

```python
# Self-test
g = Gatto('Felix')
assert g.nome == 'Felix'        # attributo ereditato
assert g.verso() == 'Miao'
assert isinstance(g, Animale)
print('OK')
print(g.nome, '->', g.verso())
```

## Esercizio 2 - `super().__init__` + override

La classe `Veicolo` ha `__init__(self, ruote)` e il metodo
`descrivi()` che restituisce `'Veicolo con N ruote'`. Scrivi la
sottoclasse `Auto` che:
- nel costruttore riceve `marca` e fissa `ruote=4` con `super().__init__(4)`,
- fa l'**override** di `descrivi()` restituendo
  `'Auto <marca> con 4 ruote'` riusando `super().descrivi()`.

```python
class Veicolo:
    def __init__(self, ruote):
        self.ruote = ruote

    def descrivi(self):
        return f'Veicolo con {self.ruote} ruote'

# TODO: completa la sottoclasse Auto
class Auto(Veicolo):
    def __init__(self, marca):
        super().__init__(4)
        self.marca = marca
    
    def descrivi(self):
        base =  super().descrivi()
        return  base.replace("Veicolo", f"Auto {self.marca}")
    



        
```

```python
# Self-test
a = Auto('Fiat')
assert a.ruote == 4
assert a.marca == 'Fiat'
assert a.descrivi() == 'Auto Fiat con 4 ruote'
print('OK')
print(a.descrivi())
```

## Esercizio 3 - `@dataclass` base

Usa il decoratore `@dataclass` per definire `Punto` con due campi
interi `x` e `y`. Non scrivere `__init__` ne' `__repr__`: li genera la
dataclass.

```python
from dataclasses import dataclass

# TODO: completa la dataclass Punto
@dataclass
class Punto:
    x: int
    y: int

p = Punto(82, 11)
print(p)
```

```python
# Self-test
p = Punto(3, 4)
assert p.x == 3 and p.y == 4
assert repr(p) == 'Punto(x=3, y=4)'   # __repr__ automatico
assert Punto(3, 4) == Punto(3, 4)      # __eq__ automatico
print('OK')
print(p)
```

## Esercizio 4 - `field(default_factory=list)`

Definisci la dataclass `Carrello` con un campo `articoli` che parte da
una **lista vuota** (usa `field(default_factory=list)`) e un metodo
`aggiungi(nome)` che appende `nome` alla lista.

```python
from dataclasses import dataclass, field

# TODO: completa la dataclass Carrello
@dataclass
class Carrello:
    articoli: list = field(default_factory=list)

    def aggiungi(self, nome):
        self.articoli.append(nome)

a = Carrello("cereali")
b = Carrello("maionese")
print(b)
```

```python
# Self-test
c1 = Carrello()
c2 = Carrello()
c1.aggiungi('mela')
assert c1.articoli == ['mela']
assert c2.articoli == []        # le liste NON sono condivise
print('OK')
print('c1:', c1.articoli, '| c2:', c2.articoli)
```

## Esercizio 5 - `@dataclass(frozen=True)`

Definisci una dataclass **immutabile** `Coordinate` con campi
`lat` e `lon` (float). Deve confrontarsi per valore e non permettere la
modifica dei campi dopo la creazione.

```python
from dataclasses import dataclass

# TODO: completa la dataclass immutabile Coordinate
@dataclass(frozen=True)
class Coordinate:
    lat: float;
    lon: float;
```

```python
# Self-test
c = Coordinate(45.07, 7.69)
assert c == Coordinate(45.07, 7.69)    # uguaglianza per valore
bloccata = False
try:
    c.lat = 0.0                        # deve essere vietato
except Exception:
    bloccata = True
assert bloccata, 'la dataclass dovrebbe essere immutabile (frozen=True)'
print('OK')
print(c)
```

## Esercizio 6 - Mini progetto: classe astratta `Esportatore`

Definisci una classe astratta `Esportatore` (eredita da `ABC`) con un
metodo astratto `esporta(self, dati)` (dove `dati` e' una lista di
stringhe). Poi scrivi la sottoclasse concreta `EsportatoreCSV` che
implementa `esporta` restituendo i dati uniti da virgole.
Esempio: `['a', 'b', 'c']` -> `'a,b,c'`.

```python
from abc import ABC, abstractmethod

# TODO: completa la classe astratta e la sottoclasse concreta
class Esportatore(ABC):
    @abstractmethod
    def esporta(self, dati: list[str]):
        return dati

class EsportatoreCSV(Esportatore):
    def esporta(self, dati):
        return ",".join(dati)

instance = EsportatoreCSV()
print(instance.esporta(['a','b','c']))
```

```python
# Self-test
vietato = False
try:
    Esportatore()                      # non si puo' istanziare una classe astratta
except TypeError:
    vietato = True
assert vietato, 'Esportatore dovrebbe essere astratta'

csv = EsportatoreCSV()
assert csv.esporta(['a', 'b', 'c']) == 'a,b,c'
assert isinstance(csv, Esportatore)
print('OK')
print(csv.esporta(['nome', 'ruolo', 'token']))
```
