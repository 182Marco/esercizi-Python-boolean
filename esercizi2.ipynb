## Esercizio 1 - `*args`

Scrivi la funzione `media(*numeri)` che restituisce la **media**
dei numeri ricevuti. Se non riceve numeri, deve restituire `0.0`.

```python
# TODO: completa la funzione usando *args
def media(*numeri: float | int) -> float:
    if not numeri: return 0.0
    return sum(numeri) / len(numeri)

media(0, 100)
```

```python
# Self-test
assert media(2, 4, 6) == 4.0
assert media(10) == 10.0
assert media() == 0.0
print("OK")
print("media(2,4,6) =", media(2, 4, 6))
```

## Esercizio 2 - `**kwargs`

Scrivi `crea_config(**opzioni)` che restituisce un dizionario con
questi default, sovrascritti dai valori passati:
`model='gpt-4o-mini'`, `temperature=0.0`, `max_tokens=512`.

```python
# TODO: usa opzioni.get(chiave, default) per ogni parametro
def crea_config(**opzioni):
    return {
    'model' : opzioni.get('model', 'gpt-4o-mini'),
    'temperature': opzioni.get('temperature', 0.0),
    'max_tokens' : opzioni.get('max_tokens', 512)
    }


crea_config(model = 'porco', temperature = 3298 )
```

```python
# Self-test
assert crea_config() == {"model": "gpt-4o-mini", "temperature": 0.0, "max_tokens": 512}
assert crea_config(temperature=0.9)["temperature"] == 0.9
assert crea_config(model="gpt-4o", max_tokens=2048)["max_tokens"] == 2048
print("OK")
print(crea_config(model='claude-opus'))
```

## Esercizio 3 - Unpacking

Data la funzione `intervallo(inizio, fine, passo)` gia' scritta,
chiamala spacchettando **prima** la lista `args_lista` con `*` e
**poi** il dizionario `args_dict` con `**`. Salva i risultati in
`r1` e `r2`.

```python
def intervallo(inizio, fine, passo):
    return list(range(inizio, fine, passo))

args_lista = [0, 10, 2]
args_dict = {"inizio": 1, "fine": 10, "passo": 3}

# TODO: chiama intervallo usando * e **
r1 = intervallo(*args_lista)
r2 = intervallo(**args_dict)
```

```python
# Self-test
assert r1 == [0, 2, 4, 6, 8]
assert r2 == [1, 4, 7]
print("OK")
print("r1 =", r1)
print("r2 =", r2)
```

## Esercizio 4 - `lambda` e `sorted`

Dato `modelli` (lista di dict con `nome` e `prezzo`), costruisci:
- `per_prezzo` -> ordinata per `prezzo` **crescente**,
- `per_nome`   -> ordinata per `nome` in ordine **alfabetico**.

Usa `sorted(..., key=lambda ...)`.

```python
modelli = [
    {"nome": "llama",       "prezzo": 0.0009},
    {"nome": "gpt-4o",      "prezzo": 0.0025},
    {"nome": "claude-opus", "prezzo": 0.0150},
]

# TODO
per_prezzo = sorted(modelli, key=lambda  m : m['prezzo'])
per_nome = sorted(modelli, key=lambda  m : m['nome'])

print(per_nome)
```

```python
# Self-test
assert [m["nome"] for m in per_prezzo] == ["llama", "gpt-4o", "claude-opus"]
assert [m["nome"] for m in per_nome] == ["claude-opus", "gpt-4o", "llama"]
print("OK")
print("per_prezzo:", [m["nome"] for m in per_prezzo])
print("per_nome:  ", [m["nome"] for m in per_nome])
```
